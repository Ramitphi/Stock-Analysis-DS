```python
def set_trend(row):
    if row["Daily per change"] >=-0.5 and row['Daily per change']<=0.5:
        return "Slight or No change"
    if row["Daily per change"] >= 0.5 and row['Daily per change']<=1:
        return "Slight Positive"
    if row["Daily per change"] >=-1 and row['Daily per change']<=-0.5:
        return "Slight Negative"
    elif row["Daily per change"] >=1 and row['Daily per change']<=3:
        return "Positive"
    elif row["Daily per change"] >=-3 and row['Daily per change']<=-1:
        return "Negative"
    elif row["Daily per change"] >= 3 and row['Daily per change']<= 7:
        return "Among Top Gainers"
    elif row["Daily per change"] >= -7 and row['Daily per change']<= -3:
        return "Among Top Losers"
    elif row["Daily per change"] >=7:
        return "Bull run"
    elif row['Daily per change']<=-7:
        return "Bear Drop"
df = df.assign(Trend=df.apply(set_trend, axis=1))

print(df)
```
